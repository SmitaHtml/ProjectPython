# Matplotlib

```python
df_10 = pd.read_sql_query("SELECT institution, world_rank FROM uni WHERE world_rank < 9  GROUP BY institution order by world_rank ", cn1)
```

```python
df_in = df_10.set_index('institution')
ax = df_in.plot.bar(stacked=True, title= "World's top University")
ax.set_xlabel("institution",fontsize=12)
ax.set_ylabel("world_rank",fontsize=12)
```

```python
df_100 = pd.read_sql_query("SELECT country, count(institution) FROM uni WHERE world_rank < 100 GROUP BY country", cn1 )
```

```python
df_in = df_100.set_index('country')
ax = df_in.plot.bar(stacked=True, title= "Top Universities by country", color='red')
ax.set_xlabel("country",fontsize=12)
ax.set_ylabel("count(institution)",fontsize=12)
```

# SQL Query

```python
df_1 = pd.read_sql_query("SELECT * FROM uni WHERE country = 'India'", cn1)

df_1
```

```python
df_2 = pd.read_sql_query("SELECT * FROM uni WHERE country IN ('Germany', 'France', 'Japan') LIMIT 5 ", cn1)
 
df_2

```

```python
df_3 = pd.read_sql_query("SELECT COUNT(institution),country FROM uni GROUP BY country ORDER BY COUNT(institution) DESC  ", cn1)
 
df_3
```

```python
df_2 = pd.read_sql_query("SELECT institution,country,quality_of_education FROM uni ORDER BY quality_of_education ASC LIMIT 25", cn1)

df_2.head()
```

```python
# SQL JOIN
join_df = pd.read_sql_query("""
SELECT   institution,num_students,international_students
FROM uni
JOIN time
ON uni.world_rank = time.world_rank
""", cn1)
join_df.head()

```

```python
# SQL UNION
year_union = pd.read_sql_query("""
SELECT year FROM uni
UNION
SELECT year FROM time
ORDER BY year;
""", cn1)
year_union.head()


```
