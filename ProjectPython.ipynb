```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sqlite3
connection = sqlite3.connect("company.db")
cn1 = sqlite3.connect("pythonproject.db")
df = pd.read_csv("WorldUniversityRankings.csv" , header='infer')
df.to_sql("uni", cn1, if_exists = "replace")

df
```

```python
df_10 = pd.read_sql_query("SELECT institution, national_rank FROM uni WHERE world_rank < 10 GROUP BY institution ", sqlite3.connect("pythonproject.db"))
```

```python
df_in = df_10.set_index('institution')
ax = df_in.plot.bar(stacked=True, title= "World's top 10 University")
ax.set_xlabel("institution",fontsize=12)
ax.set_ylabel("national_rank",fontsize=12)
```

```python
df_10 = pd.read_sql_query("SELECT country, count(institution) FROM uni WHERE world_rank < 100 GROUP BY country", sqlite3.connect("pythonproject.db"))
```

```python
df_in = df_10.set_index('country')
ax = df_in.plot.bar(stacked=True, title= "Universities in country")
ax.set_xlabel("country",fontsize=12)
ax.set_ylabel("count(institution)",fontsize=12)
```
